```python
# Installer la bibliothèque yfinance
!pip install yfinance

# Importer les bibliothèques nécessaires
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Récupérer les données des trois actions et du S&P 500
tickers = ['AAPL', 'GOOGL', 'MSFT', '^GSPC']
start_date = '2020-01-01'
end_date = '2023-03-29'

stocks_data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']

# Calculer les rendements quotidiens
daily_returns = stocks_data.pct_change()

# Calculer les rendements cumulatifs
cumulative_returns = (1 + daily_returns).cumprod()

# Comparer les rendements quotidiens
plt.figure(figsize=(14, 7))
sns.lineplot(data=daily_returns)
plt.title('Rendements quotidiens')
plt.xlabel('Date')
plt.ylabel('Rendement quotidien')
plt.show()

# Comparer les rendements cumulatifs
plt.figure(figsize=(14, 7))
sns.lineplot(data=cumulative_returns)
plt.title('Rendements cumulatifs')
plt.xlabel('Date')
plt.ylabel('Rendement cumulatif')
plt.show()

# Comparer la volatilité des actions et de l'indice S&P 500
rolling_std = daily_returns.rolling(window=30).std()
plt.figure(figsize=(14, 7))
sns.lineplot(data=rolling_std)
plt.title('Volatilité des actions et de l\'indice S&P 500 (écart-type sur 30 jours)')
plt.xlabel('Date')
plt.ylabel('Volatilité')
plt.show()

# Comparer les performances avec un portefeuille équilibré et l'indice S&P 500
weights = np.array([1/3, 1/3, 1/3, 0])
portfolio_returns = daily_returns.dot(weights)
cumulative_portfolio_returns = (1 + portfolio_returns).cumprod()

plt.figure(figsize=(14, 7))
sns.lineplot(data=cumulative_portfolio_returns, label='Portefeuille équilibré')
sns.lineplot(data=cumulative_returns)
plt.title('Comparaison des performances avec un portefeuille équilibré et l\'indice S&P 500')
plt.xlabel('Date')
plt.ylabel('Rendement cumulatif')
plt.legend()
plt.show()
```

```python
# Installer la bibliothèque yfinance
!pip install yfinance

# Importer les bibliothèques nécessaires
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Récupérer les données de LVMH et Kering
tickers = ['MC.PA', 'KER.PA']
start_date = '2020-01-01'
end_date = '2023-03-29'

stocks_data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']

# Calculer les rendements quotidiens
daily_returns = stocks_data.pct_change()

# Calculer les rendements cumulatifs
cumulative_returns = (1 + daily_returns).cumprod()

# Graphique des cours de clôture ajustés
plt.figure(figsize=(14, 7))
sns.lineplot(data=stocks_data)
plt.title('Cours de clôture ajustés')
plt.xlabel('Date')
plt.ylabel('Prix ajusté')
plt.show()

# Histogramme des rendements quotidiens
plt.figure(figsize=(14, 7))
sns.histplot(data=daily_returns, kde=True, bins=50, alpha=0.5)
plt.title('Histogramme des rendements quotidiens')
plt.xlabel('Rendement quotidien')
plt.ylabel('Fréquence')
plt.show()

# Matrice de corrélation des rendements quotidiens
corr_matrix = daily_returns.corr()

plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matrice de corrélation des rendements quotidiens')
plt.show()

# Graphique à barres des rendements cumulatifs
cumulative_returns.iloc[-1].plot(kind='bar', figsize=(10, 6))
plt.title('Rendements cumulatifs')
plt.xlabel('Actions')
plt.ylabel('Rendement cumulatif')
plt.show()
```
